```python
import numpy as np
import matplotlib.pyplot as plt
```

## Grafik çizimi

```python
%matplotlib inline  
#plt.show yazmayı sürekli gerektirmez
```

```python
yasListesi = [10, 20, 30, 30, 50, 60, 70, 75, 75]
kiloListesi = [60, 60, 75, 80, 70, 82, 74, 71, 71]

numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi, numpyKiloListesi, "r")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Yaş-kilo değişim grafiği")
# "r çiginin rengini belirer (x , y, red)"
```

```python
numpyDizisi1 = np.linspace(0 , 10, 20)
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "b*-")
```

```python
plt.subplot(1, 2, 1)
plt.plot(numpyDizisi1, numpyDizisi2, "r*-")
plt.subplot(1, 2, 2)
plt.plot(numpyDizisi2, numpyDizisi1, "b--")
```

## Figure oluşturma

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2, 0.2, 0.9, 0.9])
figureAxes.plot(numpyDizisi1, numpyDizisi2, "g--")
figureAxes.set_xlabel("X ekseni ismi")
figureAxes.set_ylabel("Y ekseni ismi")
figureAxes.set_title("Grafik başlığı")
```

```python
figure1 = plt.figure()

# [x, y, width, height]
figureAxes1 = figure1.add_axes([0.1, 0.1, 0.7, 0.7])
figureAxes2 = figure1.add_axes([0.2, 0.4, 0.3, 0.3])

figureAxes1.plot(numpyDizisi1, numpyDizisi2, "g")
figureAxes1.set_xlabel("X ekseni ismi")
figureAxes1.set_ylabel("Y ekseni ismi")
figureAxes1.set_title("Grafik başlığı")

figureAxes2.plot(numpyDizisi2, numpyDizisi1, "y")
figureAxes2.set_xlabel("X ekseni ismi")
figureAxes2.set_ylabel("Y ekseni ismi")
figureAxes2.set_title("Grafik 2 başlığı")
```

```python
(benimFigure, benimEksen) = plt.subplots(nrows=1, ncols=2)
for eksen in benimEksen:
    eksen.plot(numpyDizisi1, numpyDizisi2, "g")
    eksen.set_xlabel("X ekseni")

plt.tight_layout() #eksenler arası görsel mesafeyi ayarlıyor

#benimFigure
```

```python
yeniFigur = plt.figure(dpi = 100)
yeniEksen = yeniFigur.add_axes([0.1, 0.1, 0.9, 0.9])
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 2, "g*-", label = "numpyDizisi1 ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label = "numpyDizisi1 ** 3")
yeniEksen.legend(loc=0) #label'ı gösterir
```
