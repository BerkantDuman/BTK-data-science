```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline  
#plt.show yazmayı sürekli gerektirmez
```

```python
numpyDizisi1 = np.linspace(0 , 10, 20)
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(myFigure, myEksen) = plt.subplots()
myEksen.plot(numpyDizisi1, numpyDizisi2, color = "#C96F23", alpha = 0.3)
myEksen.plot(numpyDizisi1, color = "#3A95A8")
```

```python
(yeniFigıre, yeniEksen) = plt.subplots(dpi = 150)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, "red", linewidth = 0.8, linestyle = ":")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, "purple", linestyle = "--")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 9, color = "b", linestyle = "-.", marker ="o", markersize = 3, markerfacecolor = "yellow")
```

## Scatter

```python
plt.scatter(numpyDizisi1, numpyDizisi2)
```
